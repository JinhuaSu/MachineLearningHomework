```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def random_walking(t=1,delta_t = 0.5**(6)):
    delta_x = np.sqrt(delta_t)
    size = int(t/delta_t)
    x_series = delta_x * np.cumsum(np.random.choice([-1,1],size = size))
    t_series = [delta_t * (i+1) for i in range(size)]
    plt.figure()
    plt.plot(t_series,x_series)
```

```python
for i in range(5):
    random_walking()
```

```python
for i in range(5):
    random_walking(delta_t = 2**(-10))
```
